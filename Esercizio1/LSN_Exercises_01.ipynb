### In attempt to lower the impact of autocorrelation in a simulation data we will use in every exercise the blocking averaging, which consist in dividing the simulation values into blocks and calculating averages within each block and this will make more accurate estimates.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd

plt.rcParams.update({'font.size': 20})

M=100000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N) 
x1 = np.arange(N) 
x1*=L
```

### Exercise 01.3
- **Simulate** the Buffons experiment:  A needle of length $L$ is thrown at random onto a horizontal plane ruled with straight lines a distance $d$ (must be $d > L$, but do not use $d\gg L$ otherwise $P\ll 1$) apart. The probability $P$ that the needle will intersect one of these lines is: $P = 2L/\pi d$. This could be used to evaluate $\pi$ from throws of the needle: if the needle is thrown down $N_{thr}$ times and is observed to land on a line $N_{hit}$ of those times, we can make an estimate of $\pi$ from
$$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$
Make a picture of the estimation of $\pi$ and its uncertainty (Standard Deviation of the mean) with a large number of *throws* $M$ as a function of the number of blocks, $N$</font> (see below: Computing statistical uncertainties). If possible, do not use $\pi$ to evaluate $\pi$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd

plt.rcParams.update({'font.size': 20})
data4 = pd.read_csv("../Esercizio1/Esercizio1.3/dati.dat", sep='\t')


M=100000              # Total number of throws
N=100                # Number of blocks
L=int(M/N) 
x1 = np.arange(N) 
x1*=L
plt.errorbar(x1,(data4["sum"]-math.pi),yerr=data4["err"])
plt.rcParams["figure.figsize"]=(10,5)
plt.title('Estimation of $\pi$')
plt.axhline(y=0, color='r', linestyle='-')
plt.xlabel('#throws')
plt.ylabel('<r>-$\pi$')
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

**Question**: $N$ can be chosen in many ways ... what are the limitations regarding this choice?

### When choosing the value of $N$ in a Monte Carlo calculation, there are several limitations to consider(expect the computational ones if i use very big values of $N$ the computational performarce will be low):

### 1)Trade-off between bias and statistical uncertainty: the choice of $N$ affects the trade-off between the statistical uncertainty and bias in the estimation of the quantity of interest. A larger $N$ generally leads to smaller statistical uncertainties, but can it also introduce a more bias if the Monte Carlo algorithm has not converged or if there are systematic errors in the simulation.

### 2)Autocorrelation: In many Monte Carlo calculations, the samples generated in successive steps are not statistically independent due to correlations between them. If $N$ is chosen too small, the samples used for estimating the uncertainty may be highly correlated, leading to an underestimation of the true uncertainty.

### 3)Convergence of the Monte Carlo algorithm: The choice of $N$ should ensure that the Monte Carlo algorithm has converged to the desired level of accuracy. If $N$ is too small, the estimate may still be far from the true value, resulting in an unreliable uncertainty estimate.
